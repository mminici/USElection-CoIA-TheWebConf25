```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm.notebook import tqdm
import time
from itertools import product
from tqdm import tqdm
import os
import pickle
from pathlib import Path

import warnings
from scipy.sparse import SparseEfficiencyWarning

# Suppress the SparseEfficiencyWarning
warnings.simplefilter('ignore', SparseEfficiencyWarning)

USE_GPU = True
TEST = False

from pathlib import Path
if USE_GPU:
    import cudf
    from cuml.feature_extraction.text import TfidfVectorizer
    import cupy as cp  # To check if GPU is available
    import cudf
else:
    from sklearn.feature_extraction.text import TfidfVectorizer
    from sklearn.metrics import pairwise_distances

import networkx as nx
import polars as pl
import time
from scipy.sparse import csr_matrix
import networkx as nx
import os
import pickle
from cuml.metrics import pairwise_distances






def plot_cdf_threshold(array, name, perc_threshold, cumulative=True, output_name=None, verbose=False):
    """perc_threshold: threshold in percentile [0-100]
    """
    # Calculate percentiles and plot vertical lines
    percentile_25 = np.percentile(array, 25)
    percentile_75 = np.percentile(array, 75)
    percentile_90 = np.percentile(array, 90)
    min_value = np.percentile(array, perc_threshold)
    if not verbose:
        return min_value

    
    plt.figure(figsize=(6, 4))
    
    # Plot CDF using KDE with cumulative=True
    sns.kdeplot(array, color='skyblue', cumulative=cumulative, fill=True)
    #plt.hist(array, bins=50, color='gray', edgecolor='black', alpha=0.4, label='Histogram', density=True)

    
    plt.title(f'Cumulative Distribution of {name}')
    plt.xlabel(f'{name}')
    plt.ylabel('Cumulative Probability')
    

    plt.axvline(percentile_25, color='green', linestyle='--', 
                label=f'25th percentile: {percentile_25:.2f}')
    plt.axvline(percentile_75, color='green', linestyle='--', 
                label=f'75th percentile: {percentile_75:.2f}')
    plt.axvline(percentile_90, color='green', linestyle='--', 
                label=f'90th percentile: {percentile_90:.2f}')
    plt.axvline(min_value, color='red', linestyle='--', 
                label=f'CHOSEN {perc_threshold}th percentile: {min_value:.2f}')
    
    sns.despine()
    plt.xlim(0.)
    plt.legend(loc='lower right')
    os.makedirs(f'{output_name}/', exist_ok=True)
    if output_name:
        plt.tight_layout()
        plt.savefig(f'{output_name}/{output_name}-{name}.pdf', dpi=300, bbox_inches='tight') 
    plt.show()
    plt.close()
    
    return min_value

def plot_distribution_threshold(array, name, perc_threshold, cumulative=False, output_name=None, verbose=False):
    """perc_threshold: threshold in percentile [0-100]
    """ 
    # Calculate percentiles and plot vertical lines
    percentile_25 = np.percentile(array, 25)
    percentile_75 = np.percentile(array, 75)
    percentile_90 = np.percentile(array, 90)
    
    min_value = np.percentile(array, perc_threshold)
    if not verbose:
        return min_value

    plt.figure(figsize=(6, 4))
    plt.hist(array, bins=50, color='skyblue', edgecolor='black', alpha=0.7, cumulative=cumulative)
    plt.title(f'Distribution of {name}')
    plt.xlabel(f'{name}')
    plt.ylabel('Frequency')
    
  

    plt.axvline(percentile_25, color='green', linestyle='--', 
                label=f'25th percentile: {percentile_25:.2f}')
    plt.axvline(percentile_75, color='green', linestyle='--', 
                label=f'75th percentile: {percentile_75:.2f}')
    plt.axvline(percentile_90, color='green', linestyle='--', 
                label=f'90th percentile: {percentile_90:.2f}')
    plt.axvline(min_value, color='red', linestyle='--', 
                label=f'CHOSEN {perc_threshold}th percentile: {min_value:.2f}')
    if not cumulative:
        plt.yscale("log")
    sns.despine()
    plt.xlim(0.)
    plt.legend(loc='upper right')
    os.makedirs(f'{output_name}/', exist_ok=True)
    if output_name:
        plt.tight_layout()
        plt.savefig(f'{output_name}/{output_name}-{name}.pdf', dpi=300, bbox_inches='tight') 
    plt.show()
    plt.close()
    
    return min_value

def display_similarity_threshold(sparse_matrix_csr, perc_threshold, use_cdf=True, output_name=None, verbose=True):
    similarities = sparse_matrix_csr.data.flatten()
    if use_cdf:
        return plot_distribution_threshold(similarities, "Similarity", perc_threshold, output_name=output_name, verbose=verbose)
    return plot_distribution_threshold(similarities, "Similarity", perc_threshold, output_name=output_name, verbose=verbose)

def display_centrality_values(centrality_values, perc_threshold, use_cdf=True, output_name=None, verbose=True):
    values = list(centrality_values.values())
    if use_cdf:
        return plot_cdf_threshold(values, "Eigenvector Centrality", perc_threshold, cumulative=True, output_name=output_name, verbose=verbose)
    return plot_distribution_threshold(values, "Eigenvector Centrality", perc_threshold, cumulative=True, output_name=output_name, verbose=verbose)

def display_courl_count(similarity_graph, weight_name='value', perc_threshold=0, use_cdf=True, output_name=None, verbose=True):
    values = [edge[2][weight_name] for edge in similarity_graph.edges(data=True)]
    if use_cdf:
        return plot_cdf_threshold(values, "Co-url Count", perc_threshold, output_name=output_name, verbose=verbose)
    return plot_distribution_threshold(values, "Co-url Count", perc_threshold, output_name=output_name, verbose=verbose)





### NETWORK


import networkx as nx
from pyvis.network import Network
from scipy.sparse import find  # Use `find` to extract non-zero values from sparse matrices
from sklearn.metrics.pairwise import cosine_similarity
import polars as pl
from itertools import combinations
### VIZ Utils
import sqlite3
import pandas as pd
import json


def get_user_urls(df, user_id, label):
    # Filter the DataFrame for the specific user and get their 'urls' as a set
    urls = df.filter(pl.col(label) == user_id).select("urls").to_series()[0].split(" ")
    #urls = df.filter(pl.col(label) == int(user_id)).select("urls").to_series().to_list() # use the list
    if urls:
        # Return the set of URLs for that user
        return set(urls)
    return set()

# Function to compute the number of common URLs between two users
def common_urls_count(df, user1, user2, label):
    urls_user1 = get_user_urls(df, user1, label)
    urls_user2 = get_user_urls(df, user2, label)
    
    # Find the intersection of the two sets
    common_urls = urls_user1.intersection(urls_user2)
    
    # Return the count of common URLs
    return common_urls


# Define a function to assign node shapes and colors based on the platform (suffix)
def get_node_attributes(platform):
    if platform == "Twitter":
        return "dot", "black"  # Circle with Twitter blue color
    elif platform == "Telegram":
        return "triangle", "deepskyblue"  # Triangle with Telegram blue color
    elif platform == "Facebook":
        return "hexagon", "dodgeblue"  # Hexagon with Facebook blue color
    else:
        return "dot", "gray"  # Default: Circle with light gray color





# Loading
import polars as pl
from pathlib import Path
from tqdm.notebook import tqdm


def read_cross_dataframes(MIN_URLS_PER_USER, UNRELATED_NODES, verbose=False):
    data_path = Path(f"{main_path}/telegram-clean/channel2urls_expanded_domains.parquet")
    df_telegram = pl.read_parquet(data_path).with_columns([
        pl.col('channel_id').cast(pl.Utf8)  # Cast user_id to string
    ])
    if verbose: display(df_telegram.head(2))
    if df_telegram['channel_id'].null_count():
        print("Nan users in Telegram")
    
    data_path = Path(f"{main_path}/twitter-clean/user_id2urls_expanded_domains.parquet")
    df_twitter = pl.read_parquet(data_path).with_columns([
        pl.col('user_id').cast(pl.Utf8)  # Cast user_id to string
    ])
    if verbose: display(df_twitter.head(2))
    if df_twitter['user_id'].null_count():
        print("Nan users in Twitter")
    
    data_path = Path(f"{main_path}/facebook-clean/user_id2urls_expanded_domains.parquet")
    df_facebook = pl.read_parquet(data_path).with_columns([
        pl.col('user_id').cast(pl.Utf8)  # Cast user_id to string
    ])
    if verbose: display(df_facebook.head(2))


    # Filtering
    MIN_URLS_PER_USER = 10
    COL = "urls_expanded"
    
    df_twitter = df_twitter.filter(
        pl.col(COL).list.len() > MIN_URLS_PER_USER
    )
    
    df_telegram = df_telegram.filter(
        pl.col(COL).list.len() > MIN_URLS_PER_USER
    )
        
    df_facebook = df_facebook.filter(
        pl.col(COL).list.len() > MIN_URLS_PER_USER
    )
    
    if verbose:
        print(f"Twitter: {len(df_twitter):,}, Telegram: {len(df_telegram):,}, Facebook: {len(df_facebook):,}")


    # Combine
    # Remove unrelated users for Twitter
    if UNRELATED_NODES['twitter']:
        df_twitter = df_twitter.filter(~pl.col("user_id").is_in(UNRELATED_NODES['twitter']))
    # Add the user suffix for Twitter
    df_twitter_modified = df_twitter.with_columns(
        (pl.col("user_id").cast(pl.Utf8) + "_twitter_user").alias("user_id")
    ).select(["user_id", "profile_url", "urls", COL, "domains_expanded"])
    
    # Remove unrelated users for Telegram
    if UNRELATED_NODES['telegram']:
        df_telegram = df_telegram.filter(~pl.col("channel_id").is_in(UNRELATED_NODES['telegram']))
    # Add the user suffix for Telegram
    df_telegram_modified = df_telegram.with_columns(
        (pl.col("channel_id").cast(pl.Utf8) + "_telegram_channel").alias("user_id")
    ).select(["user_id", "profile_url", "urls", COL, "domains_expanded"])
    
    # Remove unrelated users for Facebook
    if UNRELATED_NODES['facebook']:
        df_facebook = df_facebook.filter(~pl.col("user_id").is_in(UNRELATED_NODES['facebook']))
    # Add the user suffix for Facebook
    df_facebook_modified = df_facebook.with_columns(
        (pl.col("user_id").cast(pl.Utf8) + "_facebook_page").alias("user_id")
    ).select(["user_id", "profile_url", "urls", COL, "domains_expanded"])
    
    
    # Concatenate both dataframes
    df_combined = pl.concat([df_twitter_modified, df_telegram_modified, df_facebook_modified])

    return df_combined





# Processing

def get_graph_instance(iodf, filtered_similarity_graph: nx.Graph, PLATFORM, MIN_COMMON_URLS, MIN_SIMILARITY, verbose=True):
    """Returns graph formatted for viz and filtered by MIN_COMMON_URLS, MIN_SIMILARITY
    """
    USER_LABEL = "channel_id" if PLATFORM == "telegram" else "user_id"

    platform2 = platform1 = PLATFORM
    # Step 2: Create a graph from the similarity matrices without a threshold
    G = nx.Graph()
    URL2COUNTS = {}
    
    
    if verbose: print("Updating graph instance ..")
    NUM_EDGES = filtered_similarity_graph.number_of_edges()
    for _i, edge in enumerate(filtered_similarity_graph.edges(data=True)):
        if verbose: print(f"Edges: {_i+1:,}/{NUM_EDGES:,}", end="\r")
        user1, user2, data = edge
        user1 = str(user1)
        user2 = str(user2)
        weight = data['weight']
    
        # Counting and saving common urls
        common_urls = common_urls_count(iodf, user1, user2, USER_LABEL)
        for _url in common_urls:
            if _url in URL2COUNTS: URL2COUNTS[_url] += 1
            else: URL2COUNTS[_url] = 1
        num_common_urls = len(common_urls)
        if weight > 0 and num_common_urls == 0:
            print(f"Error: num_common_urls={num_common_urls}")
            print(user1, user2, weight)
            break # Error
        
        
        if num_common_urls > MIN_COMMON_URLS:
            if weight > MIN_SIMILARITY:  # Filter edges by weight (similarity score)
                # Add platform1 user node
                if user1 not in G.nodes:
                    node_shape, node_color = get_node_attributes(platform1)
                    G.add_node(user1, platform=platform1, label=user1, shape=node_shape, color='blue')
                    
                    url = iodf.filter(pl.col(USER_LABEL) == user1)["profile_url"].item()
                    
                    if platform1 == "Telegram":
                        G.nodes[user1]["title"] = f"<a href='{url}' target='_blank'>Click to view channel</a>"
                    else:
                        G.nodes[user1]["title"] = f"<a href='{url}' target='_blank'>Click to view profile</a>"
    
                # Add platform2 user node
                if user2 not in G.nodes:
                    node_shape, node_color = get_node_attributes(platform2)
                    G.add_node(user2, platform=platform2, label=user2, shape=node_shape, color='blue')
                    
                    url = iodf.filter(pl.col(USER_LABEL) == user2)["profile_url"].item()
    
                    if platform2 == "Telegram":
                        G.nodes[user2]["title"] = f"<a href='{url}' target='_blank'>Click to view channel</a>"
                    else:
                        G.nodes[user2]["title"] = f"<a href='{url}' target='_blank'>Click to view profile</a>"
    
                # Add edge between users from platform1 and platform2
                G.add_edge(user1, user2, label=f"{num_common_urls}", value=num_common_urls)
    if verbose: print("Updating graph instance done ✅")
    
    if verbose: print(f"Number of nodes: {G.number_of_nodes():,}")
    if verbose: print(f"Number of edges: {G.number_of_edges():,}")
    if verbose: print(f"Number of connected components: {nx.number_connected_components(G)}")
    
    
    
    
    # Step 3: Filtering by num of coshared urls
    MIN_COURLS = platform2parameters[PLATFORM][TYPE_OF_NETWORK]["MIN_COURLS"]
    #display_similarity_weights(G, weight_name='value', min_courl_count_value=MIN_COURLS)
    #display_courl_count(G)
    
    G_filtered = G.copy()
    edges_to_remove = [(u, v) for u, v, d in G.edges(data=True) if d['value'] <= MIN_COURLS]
    G_filtered.remove_edges_from(edges_to_remove)
    G_filtered.remove_nodes_from(list(nx.isolates(G_filtered)))
    if verbose: print(f"Number of nodes: {G_filtered.number_of_nodes():,}")
    if verbose: print(f"Number of edges: {G_filtered.number_of_edges():,}")
    if verbose: print(f"Number of connected components: {nx.number_connected_components(G_filtered)}")


    return G_filtered


# Function to measure execution time
def log_time(task_name, start_time):
    print(f"{task_name} ✅ in {time.time() - start_time:.2f} seconds")



def get_sparse_similarity_matrix(df, USER_LABEL, entity_column_name, num_to_url, MAX_DF, MIN_DF, MIN_URLS_PER_USER, verbose=True):
    """Returns scipy csr sparse matrix of similarities without selfloops
    """
    # Step 5: TF-IDF Vectorization using cuML with a max_df percentage
    start_time = time.time()
    if verbose: print("Tfidf ..")
    start_time = time.time()  # Start timing this step
    vectorizer = TfidfVectorizer(max_df=MAX_DF, min_df=MIN_DF)
    df = df.to_pandas()
    df_cudf = cudf.DataFrame.from_pandas(df)
    tfidf_matrix = vectorizer.fit_transform(df_cudf[entity_column_name])
    if verbose: log_time("Tfidf", start_time)
    if verbose: print(f"tfidf_matrix: {tfidf_matrix.shape} | vocabulary: {len(vectorizer.vocabulary_):,}", )
    if verbose: print(f"{tfidf_matrix.shape[0]:,} {USER_LABEL}s")
    

    # Step 6: Filter non empty rows
    start_time = time.time()
    if verbose: print("Filtering tfidf rows (users) with low activity ..")
    indptr = tfidf_matrix.indptr  # Shape: (num_rows + 1,)
    non_zero_row_counts = indptr[1:] - indptr[:-1] # using indptr in csr format to get non-zero counts
    non_zero_row_indices = cp.nonzero(non_zero_row_counts >= MIN_URLS_PER_USER)[0]
    tfidf_matrix = tfidf_matrix[non_zero_row_indices]  # Filter the TF-IDF matrix
    filtered_user_ids = df_cudf[USER_LABEL].to_pandas().values[non_zero_row_indices.get()]  # Filter user IDs based
    if verbose: log_time("Filtering tfidf rows (users) with low activity", start_time)
    if verbose: print(f"{tfidf_matrix.shape[0]:,} {USER_LABEL}s")
    
    ## COLUMNS
    # Get the vocabulary mapping from the vectorizer (pandas Series or values)
    vocab_tokens = vectorizer.vocabulary_.to_pandas().values
    assert not " "  in vocab_tokens
    # Create tfidf_index2num: TF-IDF col index -> numeric token
    tfidf_index2num = {i: token for i, token in enumerate(vocab_tokens) if token in num_to_url}
    # Create num2tfidf_index: numeric token -> TF-IDF col index
    num2tfidf_index = {token: i for i, token in enumerate(vocab_tokens) if token in num_to_url}
    
    
    # Step 6: Compute pairwise cosine similarity
    start_time = time.time()
    if verbose: print("Distance computation ..")
    cosine_similarity_matrix = 1 - pairwise_distances(tfidf_matrix, metric='cosine')
    if verbose: log_time("Distance computation", start_time)
    
    
    # Step 7: Construct the graph using NetworkX
    cosine_similarity_cupy = cp.sparse.csr_matrix(cosine_similarity_matrix)
    sparse_matrix_csr = csr_matrix(cosine_similarity_cupy.get())  # Move from GPU to CPU and convert to SciPy CSR
    sparse_matrix_csr.setdiag(0)  # Removing self loops

    return sparse_matrix_csr, filtered_user_ids



# ---> MAIN FUNCTION
def get_url_coordination_network(PLATFORM, PARAMETERS, hashed_params, UNRELATED_NODES, 
                                 verbose=True, TEST=False, keep_only_cross_platform_edges=True):
    """Pipeline that returns a filtered similarity graph.
    """
    # Params --------------------------------
    MIN_URLS_PER_USER = PARAMETERS["MIN_URLS_PER_USER"]
    MIN_DF = PARAMETERS["MIN_DF"]
    MAX_DF = PARAMETERS["MAX_DF"]
    
    entity_column_name = "urls"
    USER_LABEL = "channel_id" if PLATFORM == "telegram" else "user_id"
    _suffix = "channel" if PLATFORM == "telegram" else "user_id"
    
    # Check if cuML is using GPU
    if verbose: print(f"Is CUDA available? {cp.cuda.is_available()}")
    
    # ---------------------------------------
    
    # Step 1: Load the dataset using Polars
    start_time = time.time()
    if verbose: print("Loading ..") #_expanded_domains
    if PLATFORM == "cross":
        df = read_cross_dataframes(MIN_URLS_PER_USER, UNRELATED_NODES)
    else:
        file_path = Path(f'{main_path}/{PLATFORM}-clean/{_suffix}2urls_expanded_domains.parquet')
        
        df = pl.read_parquet(file_path).with_columns([
            pl.col(USER_LABEL).cast(pl.Utf8).alias(USER_LABEL)
        ])
    if PLATFORM == "twitter":
        if verbose: print(df.explode(entity_column_name)[entity_column_name].value_counts().sort(by="count", descending=True).head(25)[entity_column_name].to_list())
        GENERIC_URLS = {'http://msn.com', 'https://t.co/', 'http://youtube.com'}
        df = df.with_columns(
            pl.col("urls").map_elements(lambda url_list: [url for url in url_list if url not in GENERIC_URLS],
                                       return_dtype=pl.List(pl.Utf8))
        )
        if verbose: print(f"Remove generic urls: {GENERIC_URLS} ✅")
    
    # Remove nodes that are not related to the topic. Found by previous iterations
    if UNRELATED_NODES[PLATFORM] and PLATFORM != "cross": # for cross the filtering is already performed
        df = df.filter(~pl.col(USER_LABEL).is_in(UNRELATED_NODES[PLATFORM]))
        if verbose: print("Removed unrelated preselected nodes ✅")
    if TEST:
        if verbose: print("Sampling for test")
        df = df.sample(250)
    if verbose: log_time("Loading", start_time)
    
    
    
    # Step 2: Map the URLs in 'entity_column_name' column to numbers cuML tfidfvectorizer does not handle urls properly
    start_time = time.time()
    if verbose: print("Hashing urls ..")
    url_list = []
    for urls in df[entity_column_name]:
        url_list.extend(urls)  # Flatten the list of lists
    
    # Get unique URLs and create a URL-to-number mapping
    unique_urls = list(set(url_list))
    url_to_num = {url: str(i) for i, url in enumerate(unique_urls)}
    num_to_url = {str(i): url for i, url in enumerate(unique_urls)}  # Reverse mapping
    
    # Save the mapping to the `tmp/` folder
    os.makedirs('../tmp', exist_ok=True)
    with open('../tmp/url_to_num.pkl', 'wb') as f:
        pickle.dump(url_to_num, f)
    if verbose: log_time("Hashing urls", start_time)
    
    
    
    # Step 3: Replace URLs in 'entity_column_name' with their corresponding numbers and concatenating
    start_time = time.time()
    if verbose: print("Mapping and concat urls ..")
    def replace_urls_with_numbers(urls):
        formatted_urls = ' '.join([str(url_to_num[url]) for url in urls]).strip()
        if formatted_urls == " ":
            raise Exception(urls)
        return formatted_urls
    
    df = df.with_columns(
        df[entity_column_name].map_elements(replace_urls_with_numbers, 
                            return_dtype=pl.Utf8).alias(entity_column_name)
    )
    if verbose: log_time("Mapping and concat urls", start_time)
    
    
    
    # STEPs 4-7: Get similarity matrix with tfidf
    start_time = time.time()    
    if verbose: print("Tfidf similarity..")
    tmp_dir = Path("../tmp")
    tmp_dir.mkdir(exist_ok=True)  # Create the tmp folder if it doesn't exist
    sparse_matrix_file = tmp_dir / f"{PLATFORM}-sparse_matrix_csr.pkl"
    filtered_user_ids_file = tmp_dir / f"{PLATFORM}-filtered_user_ids.pkl"

    # Check if the files already exist
    if sparse_matrix_file.exists() and filtered_user_ids_file.exists():
        if verbose: print(f"Loading existing similarity matrix from {sparse_matrix_file}...")
        sparse_matrix_csr = load_data(sparse_matrix_file)
        filtered_user_ids = load_data(filtered_user_ids_file)
    else:
        if verbose: print("Similarity matrix not found, running get_sparse_similarity_matrix...")
        # Call the function to compute the sparse matrix and filtered user IDs
        sparse_matrix_csr, filtered_user_ids = get_sparse_similarity_matrix(df, USER_LABEL, entity_column_name, num_to_url, MAX_DF, MIN_DF, MIN_URLS_PER_USER, verbose)
        # Save the outputs to files
        save_data(sparse_matrix_file, sparse_matrix_csr)
        save_data(filtered_user_ids_file, filtered_user_ids)
    if verbose: log_time("Tfidf similarity", start_time)
    

    if keep_only_cross_platform_edges and PLATFORM == "cross":
        if verbose: print("Removing intra platform edges ..")
        start_time = time.time()
        # Step 1: Extract suffixes from user IDs
        suffixes = np.array([user_id.split('_')[-2] for user_id in filtered_user_ids])

        # Step 2: Create a mask for user pairs that have the same suffix
        mask = np.zeros(sparse_matrix_csr.shape, dtype=bool)
        for i in range(len(suffixes)):
            for j in range(i + 1, len(suffixes)):  # Check only above diagonal (symmetry of the matrix)
                if suffixes[i] == suffixes[j]:
                    mask[i, j] = True
                    mask[j, i] = True  # Ensure symmetry
        sparse_matrix_csr[mask] = 0
        if verbose: log_time("Removing intra platform edges", start_time)

    
    
    if verbose: print(f"Number of edges before filtering by similarity: {len(sparse_matrix_csr.data):,}")
    min_similarity = display_similarity_threshold(sparse_matrix_csr, 
                                PARAMETERS["MIN_SIMILARITY_PERCENTILE"],
                                output_name=output_name,
                                verbose=verbose)
    sparse_matrix_csr = sparse_matrix_csr.multiply(sparse_matrix_csr > min_similarity)
    similarity_graph = nx.from_scipy_sparse_array(sparse_matrix_csr, create_using=nx.Graph)


    
    # Step 8: Relabeling and printing stats
    mapping = dict(enumerate(filtered_user_ids))  # Map the integer labels to user IDs
    similarity_graph = nx.relabel_nodes(similarity_graph, mapping)
    if verbose: log_time("Similarity graph", start_time)

    assert len(list(nx.selfloop_edges(similarity_graph))) == 0
    if verbose: print(f'Number of nodes: {similarity_graph.number_of_nodes()}')
    if verbose: print(f'Number of edges: {similarity_graph.number_of_edges():,}')
    if verbose: print(f"Number of connected components: {nx.number_connected_components(similarity_graph)}")
    

    
    # Step 9. Remove disconnected nodes
    if verbose: print("Remove isolates ..")
    start_time = time.time()
    similarity_graph.remove_nodes_from(list(nx.isolates(similarity_graph)))
    if verbose: print(f'Number of nodes: {similarity_graph.number_of_nodes()}')
    if verbose: print(f'Number of edges: {similarity_graph.number_of_edges():,}')
    if verbose: print(f"Number of connected components: {nx.number_connected_components(similarity_graph)}")
    if verbose: log_time("Remove isolates", start_time)


    
    # Step 10. Compute node centrality values for each component
    if verbose: print("Centrality filtering ..")
    start_time = time.time()
    if similarity_graph.number_of_nodes() > 0:
        centrality_values = nx.eigenvector_centrality(similarity_graph, max_iter=500)
        display_centrality_values(centrality_values, 
                                  PARAMETERS["MIN_CENTRALITY_PERCENTILE"],
                                  output_name=output_name,
                                 verbose=verbose)
    
        def get_users_exceeding_centrality_threshold(centrality_dict, threshold):
            return [k for k, v in centrality_dict.items() if v >= threshold]
        th = PARAMETERS["MIN_CENTRALITY_PERCENTILE"]
        iolist = get_users_exceeding_centrality_threshold(centrality_values, np.percentile(list(centrality_values.values()), th))
    else:
        iolist = []
    iodf = df.filter(pl.col(USER_LABEL).is_in(iolist))

    
    # Step 11: Filtering graph instance
    filtered_similarity_graph = similarity_graph.subgraph(iolist)
    if verbose: print(f"Number of nodes: {filtered_similarity_graph.number_of_nodes():,}")
    if verbose: print(f"Number of edges: {filtered_similarity_graph.number_of_edges():,}")
    if verbose: print(f"Number of connected components: {nx.number_connected_components(filtered_similarity_graph)}")
    if verbose: log_time("Centrality filtering", start_time)

    if not filtered_similarity_graph.number_of_nodes() == len(iodf):
        print(filtered_similarity_graph.number_of_nodes(), len(iodf))
        assert filtered_similarity_graph.number_of_nodes() == len(iodf), "Try to free tmp folder"
        import pdb; pdb.set_trace()

    return filtered_similarity_graph, iodf


from pathlib import Path

def save_data(output_path, data):
    """Utility function to save data using pickle."""
    with open(output_path, 'wb') as f:
        pickle.dump(data, f)

def load_data(output_path):
    """Utility function to load data using pickle."""
    with open(output_path, 'rb') as f:
        return pickle.load(f)


import networkx as nx
import numpy as np

def compute_component_metrics(G, iodf, unique_count_domains, COMMON_LINKS):
    metrics = {
        'relative_size': [],
        'density': [],
        'clustering_coefficient': [],
        'average_shortest_path_length': [],
        'gini_domains': [],
        'simpson_domains': []
    }
    
    # Get all connected components
    components = list(nx.connected_components(G))
    number_of_components = len(components)
    total_nodes = G.number_of_nodes()

    if 'user_id' in iodf.columns: 
        USER_LABEL = 'user_id'
    elif 'channel_id' in iodf.columns: 
        USER_LABEL = 'channel_id'
    else:
        raise Exception("no user_label")
    
    for i, component in enumerate(components):
        subgraph = G.subgraph(component)
        num_nodes = len(subgraph)
        
        # Collect the metrics
        metrics['relative_size'].append(num_nodes / total_nodes)
        metrics['density'].append(nx.density(subgraph))
        metrics['clustering_coefficient'].append(nx.average_clustering(subgraph, weight='value'))
        filtered_iodf = iodf.filter(pl.col(USER_LABEL).is_in(component))

        # Concentration in domains distribution
        value_counts = (
            filtered_iodf
            .select('domains_expanded')                  # Select the exploded column
            .explode('domains_expanded')                # Explode the column
            .filter(pl.col('domains_expanded').is_in(COMMON_LINKS))
            .get_column('domains_expanded')              # Get the series
            .value_counts()                              # Perform value counts
        )
        total_count = value_counts['count'].sum()
        probabilities = value_counts['count'] / total_count
    
        # A. GINI
        extended_probabilities = np.append(probabilities, [0] * unique_count_domains)
        gini_index = 1 - np.sum(extended_probabilities**2)
        metrics['gini_domains'].append(gini_index)

        # B. simpson
        simpson_index = np.sum(probabilities.to_numpy()**2)
        metrics['simpson_domains'].append(simpson_index)
    
    # Compute min, max, median, and mean for each metric
    summary_stats = {
        'relative_size': {
            'min': np.min(metrics['relative_size']) if len(metrics['relative_size']) > 0 else None,
            'max': np.max(metrics['relative_size']) if len(metrics['relative_size']) > 0 else None,
            'median': np.median(metrics['relative_size']) if len(metrics['relative_size']) > 0 else None,
            'mean': np.mean(metrics['relative_size']) if len(metrics['relative_size']) > 0 else None
        },
        'density': {
            'min': np.min(metrics['density']) if len(metrics['density']) > 0 else None,
            'max': np.max(metrics['density']) if len(metrics['density']) > 0 else None,
            'median': np.median(metrics['density']) if len(metrics['density']) > 0 else None,
            'mean': np.mean(metrics['density']) if len(metrics['density']) > 0 else None
        },
        'clustering_coefficient': {
            'min': np.min(metrics['clustering_coefficient']) if len(metrics['clustering_coefficient']) > 0 else None,
            'max': np.max(metrics['clustering_coefficient']) if len(metrics['clustering_coefficient']) > 0 else None,
            'median': np.median(metrics['clustering_coefficient']) if len(metrics['clustering_coefficient']) > 0 else None,
            'mean': np.mean(metrics['clustering_coefficient']) if len(metrics['clustering_coefficient']) > 0 else None
        },
        # 'average_shortest_path_length': {
        #     'min': np.min(metrics['average_shortest_path_length']) if metrics['average_shortest_path_length'] else None,
        #     'max': np.max(metrics['average_shortest_path_length']) if metrics['average_shortest_path_length'] else None,
        #     'median': np.median(metrics['average_shortest_path_length']) if metrics['average_shortest_path_length'] else None,
        #     'mean': np.mean(metrics['average_shortest_path_length']) if metrics['average_shortest_path_length'] else None
        # },
        'gini_domains': {
            'min': np.min(metrics['gini_domains']) if len(metrics['gini_domains']) > 0 else None,
            'max': np.max(metrics['gini_domains']) if len(metrics['gini_domains']) > 0 else None,
            'median': np.median(metrics['gini_domains']) if len(metrics['gini_domains']) > 0 else None,
            'mean': np.mean(metrics['gini_domains']) if len(metrics['gini_domains']) > 0 else None
            
        },
        'simpson_domains': {
            'min': np.min(metrics['simpson_domains']) if len(metrics['simpson_domains']) > 0 else None,
            'max': np.max(metrics['simpson_domains']) if len(metrics['simpson_domains']) > 0 else None,
            'median': np.median(metrics['simpson_domains']) if len(metrics['simpson_domains']) > 0 else None,
            'mean': np.mean(metrics['simpson_domains']) if len(metrics['simpson_domains']) > 0 else None
            
        },
        'network': {
            'size': total_nodes
        }
        
    }
    
    return summary_stats


from collections import Counter

def count_unique_terms(df, column_name):
    """
    Given a DataFrame and a column containing lists of strings, generate a DataFrame
    that counts the occurrence of each unique string term.

    Args:
        df (pd.DataFrame): The input DataFrame.
        column_name (str): The name of the column containing lists of strings.

    Returns:
        pd.DataFrame: A DataFrame with columns 'value' and 'count' representing
                      each unique string term and its count.
    """
    # Flatten the list of all strings in the specified column
    all_strings = [item for sublist in df[column_name] for item in sublist]
    
    # Count the occurrences of each unique term
    term_counts = Counter(all_strings)
    
    # Create a DataFrame from the counts
    result_df = pd.DataFrame(term_counts.items(), columns=['value', 'count'])
    
    return result_df

def merge_values_by_case_insensitivity(df):
    """
    Given a DataFrame with columns [value, count], merge entries where 
    the 'value' differs only by capitalization.

    Args:
        df (pd.DataFrame): The input DataFrame with columns 'value' and 'count'.

    Returns:
        pd.DataFrame: A DataFrame with merged entries, case-insensitively.
    """
    # Normalize the 'value' column to lowercase
    df['value'] = df['value'].str.lower()

    # Group by the normalized 'value' and sum the counts
    merged_df = df.groupby('value', as_index=False)['count'].sum()

    return merged_df


import sys
import io

from contextlib import contextmanager

# Context manager to temporarily suppress stdout
@contextmanager
def suppress_stdout():
    # Save the current stdout
    old_stdout = sys.stdout
    sys.stdout = io.StringIO()  # Redirect stdout to an in-memory buffer
    try:
        yield
    finally:
        # Restore stdout back to normal
        sys.stdout = old_stdout
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

save = True # <-----

# Define the metrics and stat_funcs
metrics = ['density', 'gini_domains']  # The metrics for 'min' values
stat_func_min = 'min'  # Statistic for the first two metrics

# Define the third metric and its stat_func
metric_network = 'network'
stat_func_size = 'size'

# Prepare heatmap data for each metric
heatmap_data = {}

# Load data
with open(f"{PLATFORM}-results.pkl", 'rb') as results_file:
    RESULTS = pickle.load(results_file)
with open(f"{PLATFORM}-filenames.pkl", 'rb') as results_file:
    FILENAMES = pickle.load(results_file)

FILENAMES_MTX = [['' for j, y in enumerate(X_values)] for i, x in enumerate(Y_values)]
# For 'clustering_coefficient' and 'simpson_domains' with 'min' stat_func
for metric in metrics:
    data = np.full((len(Y_values), len(X_values)), np.nan)
    for i, x in enumerate(X_values):
        for j, y in enumerate(Y_values):
            # Skip filling data if the condition is not met
            if (x, y) not in FILTERED_COMBINATIONS:
                print('skip')
                continue  # Skip this x, y combination
            # Fill data and filenames if the combination is valid
            curr_result = RESULTS[(x, y)][metric][stat_func_min]
            if curr_result:
                data[j, i] = curr_result
            FILENAMES_MTX[j][i] = FILENAMES[(x, y)]
    heatmap_data[f"{metric}_{stat_func_min}"] = data

# For 'network' metric with 'size' stat_func
data_network = np.full((len(Y_values), len(X_values)), np.nan)
for i, x in enumerate(X_values):
    for j, y in enumerate(Y_values):
        if (x, y) not in FILTERED_COMBINATIONS:
            print('skip')
            continue  # Skip this x, y combination
        curr_result = RESULTS[(x, y)][metric_network][stat_func_size]
        if curr_result:
            data_network[j, i] = curr_result
        #print(x, y,  RESULTS[(x, y)][metric_network][stat_func_size])
heatmap_data[f"{metric_network}_{stat_func_size}"] = data_network

# Plotting the three heatmaps
plt.figure(figsize=(22, 6))  # Adjust figure size for 3 subplots in one row
titles = ['Minimum Density', 'Minimum Gini Index', 'Network Size']
cmaps = ["coolwarm", "coolwarm", "Blues"]
for idx, (metric_stat, data) in enumerate(heatmap_data.items()):
    plt.subplot(1, 3, idx + 1)  # 1 row, 3 columns for the three heatmaps
    sns.heatmap(data, annot=True, annot_kws={"size": 8}, 
                xticklabels=X_values, yticklabels=Y_values,
                cmap=cmaps[idx], cbar=True, linewidths=0.005,
                # vmin=0 if idx!=2 else None, 
                # vmax=.5 if idx!=2 else None
               )
    plt.title(titles[idx])
    plt.xlabel('Similarity Percentile')
    plt.ylabel('Centrality Percentile')
    plt.gca().invert_yaxis()

plt.tight_layout()

# Save the figure if output_name is provided
if output_name and save:
    plt.savefig(f'./{PLATFORM}-grid-search-heatmap.pdf', dpi=300, bbox_inches='tight')

plt.show()




# ----------------------------------
# Create an interactive HTML table
html_content = '<html><body><table border="1">'
for metric_stat, data in heatmap_data.items():
    html_content += f'<tr><th colspan="{data.shape[1]}">{metric_stat}</th></tr>'
    for i in range(data.shape[0]-1, -1, -1):  # Reverse the rows here
        html_content += '<tr>'
        for j in range(data.shape[1]):
            if data[i, j] == -777:
                continue  # Skip this x, y combination
            # Create a link to the corresponding local HTML file for each entry
            #print(data[i, j], FILENAMES_MTX[i][j], flush=True)
            link = f'{FILENAMES_MTX[i][j]}.html'  # Change to your actual paths
            html_content += f'<td><a href="{link}">{data[i, j]:.2f}</a></td>'
        html_content += '</tr>'
html_content += '</table></body></html>'

# Save the HTML table to a file
with open(f'./{PLATFORM}-grid-search-heatmap-with-links.html', "w") as f:
    f.write(html_content)

import shutil

# Zips the folder located at '../grid-search' and creates a zip file named 'grid-search.zip'
shutil.make_archive('../grid-search', 'zip', '../grid-search')
```
